```python
import sys
sys.path.append("../src")
from data_access import load_features

df = load_features([
    "default_next_month",
    "limit_bal",
    "pay_0",
    "utilization_ratio",
    "payment_ratio"]
)
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(data =df, x = "default_next_month", y = "pay_0")
plt.title("Recent delinquency (PAY_0) by default status")
plt.show()
```

```python
sns.kdeplot(
    data = df,
    x = "utilization_ratio",
    hue = "default_next_month",
    common_norm = False)
plt.title("Credit utilization by default status")
plt.show()
    
```

```python
sns.boxplot(data = df,  x = "default_next_month", y = "payment_ratio")
plt.ylim(0, 1.5)
plt.title("payment to bill ratio by default status")
plt.show()
```

```python
#Recent delinquency (pay_0) provides the strongest standalone signal of default risk. 
#Credit utilization reflects financial stress but shows substantial overlap between groups. 
#Payment-to-bill ratio strongly differentiates defaulters, with most defaulters making little or no payment, 
#while non-defaulters exhibit higher and more variable payment behavior. Default risk appears to emerge from the interaction of stress,
#behavior, and payment response rather than any single factor.
```
